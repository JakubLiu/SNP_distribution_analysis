```python
import pandas as pd
import numpy as np

human_chr1_introns_ENSG00000266602 = pd.read_csv('C:/Users/Lenovo/Desktop/STUDIA/BIOINFORMATYKA/MASTER/HUMAN_CHR1_INTRONS_GENES/human_chr1_introns_ENSG00000266602.out', sep = '\t')

#print(human_chr1_introns_ENSG00000266602.head(50))
```

```python
human_chr1_introns_ENSG00000266602.columns = ['col1', 'col2', 'col3', 'col4', 'col5', 'col6', 'col7', 'col8', 'col9', 'col10', 'col11', 'col12', 'col13', 'column_of_interest']
```

#### extracrting information about intronal SNP location

```python
list_of_interest = human_chr1_introns_ENSG00000266602['column_of_interest'].to_list()

introns_raw = []

for element in list_of_interest:
    new_element = element.split(';')
    introns_raw.append(new_element[2])
    


introns = []

for element in introns_raw:
    new_element = element.split('=')
    introns.append(new_element[1])
```

```python
introns_loc = []

for element in introns:
    new_element = element.split('/')
    introns_loc.append(new_element)
```

```python
introns_loc2 = []

n = 0

for j in introns_loc:
    introns_loc2.append(introns_loc[n][1])
    n = n + 1
    

unique_values = set(introns_loc2)
unique_list = list(unique_values)

unique_list2 = []

for i in unique_list:
    j = int(i)
    unique_list2.append(j)

unique_list2.sort()
print(unique_list2)
```

```python
# creating empty lists
intron2 = []
intron3 = []
intron4 = []
intron5 = []
intron6 = []
```

```python
y = 0

for i in introns_loc:
        
    if introns_loc[y][1] == '2':
        intron2.append(i)
        
    elif introns_loc[y][1] == '3':
        intron3.append(i)
        
    elif introns_loc[y][1] == '4':
        intron4.append(i)
        
    elif introns_loc[y][1] == '5':
        intron5.append(i)
        
    else:
        intron6.append(i)
        
    y = y + 1
```

```python
intron1_2 = []
intron2_2 = []
y = 0

for i in intron2:
    if intron2[y][0] == '1':
        intron1_2.append(i)
        
    else:
        intron2_2.append(i)
        
    y = y + 1




intron1_3 = []
intron2_3 = []
intron3_3 = []
y = 0

for i in intron3:
    
    position = intron3[y][0]
    
    if position == '1':
        intron1_3.append(i)
        
    elif position == '2':
        intron2_3.append(i)
        
    else:
        intron3_3.append(i)
        
    y = y + 1
intron1_4 = []
intron2_4 = []
intron3_4 = []
intron4_4 = []
y = 0

for i in intron4:
    
    position = intron4[y][0]
    
    if position == '1':
        intron1_4.append(i)
        
    elif position == '2':
        intron2_4.append(i)
        
    elif position == '3':
        intron3_4.append(i)
    
    else:
        intron4_4.append(i)
        
    y = y + 1
intron1_5 = []
intron2_5 = []
intron3_5 = []
intron4_5 = []
intron5_5 = []
y = 0

for i in intron5:
    
    position = intron5[y][0]
    
    if position == '1':
        intron1_5.append(i)
        
    elif position == '2':
        intron2_5.append(i)
        
    elif position == '3':
        intron3_5.append(i)
    
    elif position == '4':
        intron4_5.append(i)
    
    else:
        intron5_5.append(i)
        
    y = y + 1
intron1_6 = []
intron2_6 = []
intron3_6 = []
intron4_6 = []
intron5_6 = []
intron6_6 = []
y = 0

for i in intron6:
    
    position = intron6[y][0]
    
    if position == '1':
        intron1_6.append(i)
        
    elif position == '2':
        intron2_6.append(i)
        
    elif position == '3':
        intron3_6.append(i)
    
    elif position == '4':
        intron4_6.append(i)
    
    elif position == '5':
        intron5_6.append(i)

    else:
        intron6_6.append(i)
        
    y = y + 1
    
```

```python
# creating x-axis
two_introns_x = ['1/2', '2/2']
three_introns_x = ['1/3', '2/3', '3/3']
four_introns_x = ['1/4', '2/4', '3/4', '4/4']
five_introns_x = ['1/5', '2/5', '3/5', '4/5', '5/5']
six_introns_x = ['1/6', '2/6', '3/6', '4/6', '5/6' ,'6/6']
```

```python
two_introns_y = [len(intron1_2), len(intron2_2)]
three_introns_y = [len(intron1_3), len(intron2_3), len(intron3_3)]
four_introns_y = [len(intron1_4), len(intron2_4), len(intron3_4), len(intron4_4)]
five_introns_y = [len(intron1_5), len(intron2_5), len(intron3_5), len(intron4_5), len(intron5_5)]
six_introns_y = [len(intron1_6), len(intron2_6), len(intron3_6), len(intron4_6), len(intron5_6), len(intron6_6)]
```
